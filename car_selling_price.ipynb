```python
preprocess_var = ['mileage',	'engine',	'max_power', 'torque']
cars[preprocess_var].head(3)
```

```python
torque_list = cars['torque'].to_list()
torque_rpm = []
def extractRpm(torq_li):
  for x in torq_li:
    res = x.replace(",", "")
    temp = [int(s) for s in re.findall(r'\d+', res)]
    torque_rpm.append(max(temp))

extractRpm(torque_list)
cars['torque_rpm'] = torque_rpm
cars.head(2)
```

```python
def to_list(series):
  lis = series.to_list()
  return lis

def extractNum(series):
  ex_list = to_list(series)
  li_ex = []
  for x in ex_list:
    x_v = x.split(" ")
    li_ex.append(x_v[0])
  return li_ex

mil_list = extractNum(cars['mileage'])
eng_list = extractNum(cars['engine'])
power_list = extractNum(cars['max_power'])
cars['mileage_kmpl'] = mil_list
cars['engine_cc'] = eng_list
cars['max_power_bhp'] = power_list

cars_new = cars.drop(preprocess_var, axis = 1)
```

```python
cars_new.head(3)
```

```python
cars_new.describe()
```

```python
cars_new.info()
```

```python
# Convert categorical to integer format
cate_cols = ['fuel', 'seller_type', 'transmission', 'owner']
for i in cate_cols:
  print(cars_new[i].unique())
```

```python
# Encoding of categorical variables
def ref1(x):
  if x == 'Manual':
    return 1
  else:
    return 0

cars_new['transmission'] = cars_new['transmission'].map(ref1)

def ref2(x):
  if x == 'Individual':
    return 1
  elif x == 'Dealer':
    return 0
  else:
    return -1

cars_new['seller_type'] = cars_new['seller_type'].map(ref2)

def ref3(x):
  if x == 'Petrol':
    return 1
  elif x == 'Diesel':
    return 0
  else:
    return -1

cars_new['fuel'] = cars_new['fuel'].map(ref3)

owners = pd.get_dummies(cars_new['owner'])

X = pd.concat([cars_new, owners], axis=1)
```

```python
X.head(8)
```

```python
y = X['selling_price']
X = X.drop(['selling_price', 'name', 'owner'], axis = 1)
```

```python
#Modelling
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.decomposition import PCA
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X[:3000], y[:3000], test_size=0.2)
```

```python
Xtrain.head(2)
```

```python
random_model = RandomForestRegressor(n_estimators=300, random_state = 42, n_jobs = -1)
#fit
random_model.fit(Xtrain, ytrain)
```

```python
Xtest.head(2)
```

```python
y_pred = random_model.predict(Xtest)

#Checking the accuracy
random_model_accuracy = round(random_model.score(Xtrain, ytrain)*100,2)
print(round(random_model_accuracy, 2), '%')
```
